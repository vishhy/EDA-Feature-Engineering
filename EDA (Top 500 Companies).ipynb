## Check Multicollinearity for Categorical features
A chi-squared test (also chi-square or χ2 test) is a statistical hypothesis test that is valid to perform when the test statistic is chi-squared distributed under the null hypothesis, specifically Pearson's chi-squared test

A chi-square statistic is one way to show a relationship between two categorical variables.

Here we test correlation of Categorical columns with Target column i.e Attrition

```python
chi2_contingency = stats.chi2_contingency
hypo_result = []
for i in cat_cols:
    if chi2_contingency(pd.crosstab(data_hr['Attrition'], data_hr[i]))[1] < 0.05:   # Returns a list, at 1st index lies p-value
        hypo_result.append('Reject Null Hypothesis')
    else:
        hypo_result.append('Fail to Reject Null Hypothesis')

result = pd.DataFrame({'Variable Name':cat_cols, 'Hypothesis Result':hypo_result})
result
```
